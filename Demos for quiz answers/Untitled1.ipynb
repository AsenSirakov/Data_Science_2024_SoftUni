```python
import pandas as pd

# Load the red wine dataset
red_wine_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
red_wine = pd.read_csv(red_wine_url, sep=';')

# Calculate the correlation between alcohol content and quality
correlation = red_wine['alcohol'].corr(red_wine['quality'])

# Round the correlation to two decimal places
correlation_rounded = round(correlation, 2)

print(f"The correlation between alcohol content and quality for red wines is: {correlation_rounded}")
```

```python
url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv"
nuclear_data = pd.read_csv(url)

# Display the first few rows to understand the structure of the data
print(nuclear_data.head())

# Calculate the mean cost of a power plant in millions of USD
# Assuming the column representing the cost is named 'cost', you may need to adjust based on the dataset
mean_cost = nuclear_data['cost'].mean()

# Print the mean cost rounded to 2 decimal places
print(f"Mean cost of a power plant: {mean_cost:.2f} million USD")
```

```python
from PIL import Image
import numpy as np
from skimage import morphology
import matplotlib.pyplot as plt
```

```python
# Load the image
image = Image.open("Untitled.jpg")

# Convert image to numpy array
image_array = np.array(image)

# Calculate the mean for each channel (R, G, B)
mean_red = np.mean(image_array[:, :, 0])
mean_green = np.mean(image_array[:, :, 1])
mean_blue = np.mean(image_array[:, :, 2])

# Find the dominant channel
dominant_channel_mean = max(mean_red, mean_green, mean_blue)

# Print the result rounded to two decimal places
print(f"Mean brightness of the dominant channel: {dominant_channel_mean:.2f}")
```

```python
# Step 2: Convert the image to a NumPy array and extract the blue channel
image_array = np.array(image)
blue_channel = image_array[:, :, 2]  # Index 2 for the blue channel

# Step 3: Convert to a binary image (thresholding)
threshold_value = np.mean(blue_channel)
binary_image = blue_channel > threshold_value

# Step 4: Perform binary opening using a 3x3 square structuring element
structuring_element = morphology.square(3)
opened_image = morphology.binary_opening(binary_image, structuring_element)

# Step 5: Count the number of white pixels in the resulting image
num_white_pixels = np.sum(opened_image)

print(f"Number of white pixels after binary opening: {num_white_pixels}")

# Optionally, display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Binary Image")
plt.imshow(binary_image, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("After Binary Opening")
plt.imshow(opened_image, cmap='gray')
plt.show()
```

```python
import requests
import re
from collections import Counter
from nltk.corpus import stopwords
import nltk

# Download the text
url = "http://www.gutenberg.org/files/1342/1342-0.txt"
response = requests.get(url)
text = response.text

# Convert text to lowercase and split by non-word characters
words = re.findall(r'\b\w+\b', text.lower())

# Download stopwords if not already downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in words if word not in stop_words]

# Count the frequency of each word
word_counts = Counter(filtered_words)

# Display the most common words
most_common = word_counts.most_common(20)
print(most_common)
```

```python
# Convert text to original case format and split by non-word characters
words_with_case = re.findall(r'\b\w+\b', text)

# Remove stopwords
filtered_words_with_case = [word for word in words_with_case if word.lower() not in stop_words]

# Count the frequency of each word
word_counts_with_case = Counter(filtered_words_with_case)

# Display the most common proper nouns
most_common_proper_nouns = [word for word, count in word_counts_with_case.most_common() if word.istitle()]

print(most_common_proper_nouns[:10])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the Boston housing dataset from the original source
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Process the data to create feature matrix X and target vector y
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Convert to DataFrame for easier manipulation
columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]
boston_df = pd.DataFrame(data, columns=columns)
boston_df['MEDV'] = target

# Split the data into features and target
X = boston_df.drop(columns="MEDV")
y = boston_df["MEDV"]

# Initialize the model
model = LinearRegression()

# Fit the model on the entire dataset
model.fit(X, y)

# Compute the R^2 score
r_squared = model.score(X, y)
print(f"Coefficient of determination (R^2) for this model: {r_squared:.2f}")

# Get the coefficient related to the number of rooms (RM)
rm_index = X.columns.get_loc("RM")
coefficient = model.coef_[rm_index]
print(f"Coefficient related to the number of rooms (RM): {coefficient:.2f}")

# Get the intercept (price with all variables set to zero)
intercept = model.intercept_
print(f"Price of a hypothetical house with all variables set to zero: {intercept:.2f}")
```
