```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://github.com/plotly/datasets/raw/master/motor_trend_car_road_tests.csv'
data = pd.read_csv(url)

# Plot horsepower vs. mpg
sns.scatterplot(x='hp', y='mpg', data=data)
plt.title('Horsepower vs. MPG')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()

# Calculate correlation
hp_mpg_corr = data[['hp', 'mpg']].corr().iloc[0, 1]
print(f"Correlation between horsepower and MPG: {hp_mpg_corr:.2f}")
```

```python
sns.scatterplot(x='wt', y='qsec', data=data)
plt.title('Weight vs. 1/4 Mile Time')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('1/4 Mile Time (seconds)')
plt.show()

# Calculate correlation
wt_qsec_corr = data[['wt', 'qsec']].corr().iloc[0, 1]
print(f"Correlation between weight and 1/4 mile time: {wt_qsec_corr:.2f}")
```
