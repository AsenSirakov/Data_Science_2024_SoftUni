```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch data from the API
url = "https://www.cheapshark.com/api/1.0/deals?storeID=1&upperPrice=15"
response = requests.get(url)
data = response.json()

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())

# Basic exploration
print(f"Number of deals: {len(df)}")
print(f"Columns: {df.columns}")

# Analyze discounts
df['discountPercent'] = df['savings'].astype(float)
print(f"Average discount: {df['discountPercent'].mean():.2f}%")

# Visualize the data
plt.figure(figsize=(10, 6))
sns.histplot(df['discountPercent'], kde=True, bins=20)
plt.title('Distribution of Discounts')
plt.xlabel('Discount Percent')
plt.ylabel('Number of Deals')
plt.show()

# Top 10 deals with the highest discount
top_deals = df.sort_values(by='discountPercent', ascending=False).head(10)
print("Top 10 deals with the highest discount:")
print(top_deals[['title', 'normalPrice', 'salePrice', 'discountPercent']])

# Visualize top 10 deals
plt.figure(figsize=(12, 8))
sns.barplot(x='discountPercent', y='title', data=top_deals, palette='viridis')
plt.title('Top 10 Deals by Discount Percent')
plt.xlabel('Discount Percent')
plt.ylabel('Deal Title')
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch data from the API
url = "https://www.cheapshark.com/api/1.0/deals?storeID=1&upperPrice=15"
response = requests.get(url)
data = response.json()

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Select relevant columns and rename them for better readability
results = pd.DataFrame({
    'Title': df['title'],
    'Normal Price': df['normalPrice'],
    'Sale Price': df['salePrice'],
    'Discount Percent': df['savings'].astype(float),
    'Deal Rating': df['dealRating'],
    'Metacritic Score': df['metacriticScore'],
    'Steam Rating': df['steamRatingText']
})

# Display the DataFrame in a tabular format
print(results.head(10))

# Basic exploration
print(f"Number of deals: {len(df)}")
print(f"Columns: {df.columns}")

# Analyze discounts
average_discount = results['Discount Percent'].mean()
print(f"Average discount: {average_discount:.2f}%")

# Visualize the data
plt.figure(figsize=(10, 6))
sns.histplot(results['Discount Percent'], kde=True, bins=20)
plt.title('Distribution of Discounts')
plt.xlabel('Discount Percent')
plt.ylabel('Number of Deals')
plt.show()

# Top 10 deals with the highest discount
top_deals = results.sort_values(by='Discount Percent', ascending=False).head(10)
print("Top 10 deals with the highest discount:")
print(top_deals)

# Visualize top 10 deals
plt.figure(figsize=(12, 8))
sns.barplot(x='Discount Percent', y='Title', data=top_deals, palette='viridis')
plt.title('Top 10 Deals by Discount Percent')
plt.xlabel('Discount Percent')
plt.ylabel('Deal Title')
plt.show()
```
