```python
import pandas as pd
import folium
Green = pd.read_excel("Data/green_tripdata_2015-09.xls")

# Create a map centered around NYC
map_center = [40.7128, -74.0060]
trip_map = folium.Map(location=map_center, zoom_start=12)

# Add pickup points to the map
for _, row in Green.iterrows():
    folium.CircleMarker(
        location=(row['Pickup_latitude'], row['Pickup_longitude']),
        radius=1,
        color='blue',
        fill=True,
        fill_opacity=0.5
    ).add_to(trip_map)

# Add dropoff points to the map
for _, row in Green.iterrows():
    folium.CircleMarker(
        location=(row['Dropoff_latitude'], row['Dropoff_longitude']),
        radius=1,
        color='red',
        fill=True,
        fill_opacity=0.5
    ).add_to(trip_map)

# Save the map to an HTML file
trip_map.save('nyc_trip_map.html')
```

```python
vendor_comparison = Green.groupby('VendorID').agg(
    total_trips=('VendorID', 'size'),
    average_fare=('Fare_amount', 'mean'),
    average_trip_distance=('Trip_distance', 'mean')
).reset_index()

# Print the comparison
print(vendor_comparison)
```
